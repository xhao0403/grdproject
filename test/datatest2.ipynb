```python
newdir = 'test/'
for f in files:
    usecols = [' Label']
    labelcol = pd.read_csv(filedir+f, usecols = usecols)
    index = 0
    loop = True
    chunks = []
    chunksize = 100000
    data = pd.read_csv(filedir+f, iterator = True)
    labels = set(labelcol[' Label'])
    while loop:
        try:
            chunk = data.get_chunk(chunksize)
            for label in labels:
                tdata = chunk[chunk[' Label'] == label]
                if tdata.empty == False:
                    tdata.to_csv(newdir+label+'_data.csv', mode='a+')
                print("add data to "+newdir+str(label)+'_data.csv')
            print(f+'\' '+str(index)+"th loop")
            index += 1
        except StopIteration:
            loop = False
            print("Interation is stopped")
```
